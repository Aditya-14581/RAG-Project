```python
# Step 1: Import necessary libraries
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader
from llama_index.llms.huggingface import HuggingFaceLLM
from llama_index.core import Settings
from IPython.display import Markdown, display
import torch
from transformers import BitsAndBytesConfig
from llama_index.core import download_loader
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
```

```python
# Step 2: Scrape the Wikipedia page
URL = "https://en.wikipedia.org/wiki/Luke_Skywalker"
BeautifulSoupWebReader = download_loader("BeautifulSoupWebReader")
loader = BeautifulSoupWebReader()

# Load the data from the URL
documents = loader.load_data(urls=[URL])
```

```python
# Step 3: Configure HuggingFace LLM and quantization settings
quantization_config = BitsAndBytesConfig(
    load_in_4bit=True,                   # Use 4-bit quantization
    bnb_4bit_compute_dtype=torch.float16, # Use float16 for better performance
    bnb_4bit_quant_type="nf4",           # Use non-float 4-bit quantization
    bnb_4bit_use_double_quant=True,      # Enable double quantization for further optimization
)

# Initialize the LLM
Settings.llm = HuggingFaceLLM(
    model_name="berkeley-nest/Starling-LM-7B-alpha",
    tokenizer_name="berkeley-nest/Starling-LM-7B-alpha",
    context_window=3900,
    max_new_tokens=256,
    model_kwargs={"quantization_config": quantization_config},
    generate_kwargs={"temperature": 0.8, "do_sample": True},
    device_map="auto",
)
```

```python
# Step 4: Set up embedding model
Settings.embed_model = HuggingFaceEmbedding(model_name="BAAI/bge-small-en-v1.5")
```

```python
# Step 5: Create a VectorStoreIndex for chunking and storing data
Settings.chunk_size = 512  # Max chunk size
Settings.chunk_overlap = 50  # Overlap between chunks

# Create the vector index from the documents
vector_index = VectorStoreIndex.from_documents(documents)
```

```python
# Step 6: Create a query engine to retrieve the top 3 relevant chunks
query_engine = vector_index.as_query_engine(similarity_top_k=3)
```

```python
# Step 7: Get user input and query the engine
text = input("Input your query here about Luke Skywalker: ")

# Query the engine and get the response
response = query_engine.query(text)

# Display the response
display(Markdown(f"**Answer:** {response}"))
```
